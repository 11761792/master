# Gradual subsamples with TPOT
TPOT takes a lot of time to evaluate machine learning pipelines.
A pipeline could be faster evaluated on only a subset of the data.
The downside of this is that the resulting score of the pipeline might not reflect the true score it would get by using all available data.
There is an idea to balance the two out: evaluate new pipelines on a subset of the data, and if it proves to perform well and is taken to a next generation, increase the subset of data the pipeline will be tested on.


This approach is implemented in TPOT branch [Gradual-TPOT](https://github.com/PG-TUe/tpot/tree/a9383208d47eda9bdade6e6340756c24c685ce31).
In implementation aspect, the changes are as follows:
* each individual is evaluated only on a subset of the data initially, whenever it survives a generation, the next generation it will be evaluated on a greater subset, until eventually an indivual is evaluated on the full dataset.
* Two different ways to take subsets of the data are evaluated: 
    * "Linear", which takes a fraction 0.25, 0.5 and 1.0 of the data (used to include 0.75 as well, so linear increase of size, hence the name).
    * "Exponential", which takes a subset of 2^i instances, where `i=(9+2*gens_survived)` with `gens_survived` being the amount of generations the individual has survived (i.e. 512, 2048, ...). The data will not be oversampled, so at most every instance is in the sample exactly once.
    
This means more work is done for those individuals who survive generations, but for all other individuals less work is done.
Because there should be relatively few individuals carried over to the next generation without mutation or cross-over, this should result in an overall faster TPOT.
One can still opt-out of using this gradual evaluation by specifying so upon constructing the TPOT object.

# Quantifying the speed-up
The first task will be to evaluate if there actually was a speed-up.
Because total time taken to run TPOT varies wildly from dataset to dataset, comparing absolute times does not give a clear image.
Intead, we will evalute the speed-up factor, how many times faster (or slower) `method 1` is compared to `method 2`.

To make comparing easier, when comparing `method 1` vs `method 2`, when `method 1` is faster, it is shown in red, when `method 2` is faster, it is shown in blue.
For example, in the graph "None vs Lin", if there is a red bar with value 3.5, this means that not using GTPOT ("none") was 3.5 times faster than using Linear GTPOT ("Lin").
The average speed-up results:

```python
overall = results.groupby(by=['gradual', 'dataset']).mean()
  
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(9,3))
comparisons = [('False', 'Lin'), ('False', 'Exp'), ('Exp', 'Lin')]
mapping = {'False': 'None', 'Exp' : 'Exp', 'Lin':'Lin'}
for i, (old, new) in enumerate(comparisons):
    time_difference = (overall.loc[old].time / overall.loc[new].time).sort_values()[1:-1].reset_index().time
    speed_up = time_difference[time_difference >= 1]
    # It is easier to compare sizes of bars if they both steer on the same side of the x-axis,
    # so we make slowdowns a positive number, but will plot them in red.
    speed_down = 1 / time_difference[time_difference < 1]
    axes[i].bar(speed_down.index, speed_down, color='r')
    axes[i].bar(speed_up.index, speed_up)
    axes[i].set_title('{} vs {}'.format(mapping[old], mapping[new]))
    axes[i].set_ylabel('time factor')
    axes[i].set_xlabel('datasets')
    axes[i].tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom='off',      # ticks along the bottom edge are off
        top='off',         # ticks along the top edge are off
        labelbottom='off') 
plt.tight_layout()
```

Neither method is faster than None in *all* cases, but the slow-down of Exponential is very minor (less than 1.5 in all cases), whereas more than half of the speed-ups are at least a factor 1.5.

Next, we will inspect whether or not dataset size correlates with the speed-up Gradual-TPOT achieves, we look at dataset size in terms of samples *n* and features *d*.
The x-axis on the plots indicate the *ordering* of datasets with respect to the metric, and does not represent the value of the metric directly (for those interested, this can be found at the very bottom of this notebook, but I found it harder to read those charts, and couldn't get plotting to co-operate nicely)!

```python
from pandas import Series
from math import log10

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(9,3))

dataset_sizes = results[['dataset', 'n_instances', 'n_features']].groupby('dataset').mean().reset_index().copy()
dataset_sizes['n*d'] = Series([row['n_instances']*row['n_features'] for (_, row) in dataset_sizes.iterrows()], index=dataset_sizes.index)
dataset_sizes['n*log(d)'] = Series([row['n_instances']*log10(row['n_features']) for (_, row) in dataset_sizes.iterrows()], index=dataset_sizes.index)

gradual_methods = ['Lin', 'Exp']
dataset_size_metrics = ['n_instances', 'n_features', 'n*log(d)']

for i, gradual_method in enumerate(gradual_methods):
    for j, metric in enumerate(dataset_size_metrics):        
        time_differences = (overall.loc['False'].time / overall.loc[gradual_method].time).sort_values()[1:-1].to_frame()
        time_dif_df = dataset_sizes.join(time_differences, on='dataset')
        time_dif_df = time_dif_df.dropna()
       
        #speed_up = time_dif_df[time_dif_df >= 1]
        # It is easier to compare sizes of bars if they both steer on the same side of the x-axis,
        # so we make slowdowns a positive number, but will plot them in red.
        #speed_down = 1 / time_difference[time_difference < 1]
        #axes[i].bar(speed_down.index, speed_down, color='r')
        time_dif_df = time_dif_df.sort_values(metric).reset_index()
        axes[i,j].bar(time_dif_df.index, time_dif_df.time)
        axes[i,j].set_title('{} speed up by {}'.format(gradual_method, metric))
        axes[i,j].set_ylabel('time factor')
        axes[i,j].set_xlabel('orderd by ' + metric)
        #axes[i,j].set_xscale('log')
        axes[i,j].tick_params(
            axis='x',          # changes apply to the x-axis
            which='both',      # both major and minor ticks are affected
            bottom='off',      # ticks along the bottom edge are off
            top='off',         # ticks along the top edge are off
            labelbottom='off') 
plt.tight_layout()
```

It seems that for the Exponential method the amount of instances plays more of a role than the Linear method.
This makes sense, because the Exponential method determines the subset size without considering the full dataset size, so as the full dataset grows in size, individuals are (initially) tested on relatively smaller subsets of the data.

Looking at the figures for Exponential TPOT for both the number of instances as well as the `n*log(d)` metric, it looks like there is a fairly clear cut-off point where slow-downs stop and speed-ups start!
